# Primeiro Modelo: Multinomial Naive-Bayes


> O classificador Multinominal Naive Bayes é um método de aprendizagem probabilística que se baseia no Teorema de Bayes para prever uma classificação. Para fazer isso, ele recorre a frequência das palavras em uma base de dados teste, de forma a estipular qual a probabilidade de um elemento analisado ser de uma categoria, com base nas palavras contidas dentro deste novo elemento. O que diferencia um Multinominal Naive Bayes de um Naive Bayes convencional, é que o primeiro verifica a probabilidade em diferentes features, obtendo a classificação com base nas probabilidades calculadas em cada uma dessas features.

```python
classificador_nb = naive_bayes.MultinomialNB() # funcão Multinomial NB na biblioteca sklearn
```

# Terceiro Modelo: Método da Floresta Aleatória

>Antes de chegar na explicação do método da Floresta Aleatória iremos explicar o que é uma árvore de decisão. Esse tipo de análise se baseia responder uma pergunta e que gera ramificações na árvore baseado na resposta para a pergunta em questão, por exemplo. Em programação, consistiria em um conjunto de declarações if-else, e que quando aplicados a uma amostra de dados, resulta em sua classificação, pois a base de dados vai se ramificando de acordo com a condição relativa a cada nó da árvore. Nela, o nó raiz é o atributo que melhor divide o conjunto de dados. E os nós de decisão são representados como condições lógicas que determinam o caminho dentro da árvore. Um exemplo de árvore de decisão pode ser encontrado abaixo:

> Primeiramente, o método da floresta aleatória é um método ensemble, ou seja, é um método que combina diversos modelos buscando obter um algoritimo mais robusto. Agora sabendo o que é árvore de decisão, é possível explicar o método da floresta aleatória como uma combinação de árvores de decisão, já que esse método junta diversos árvores de decisão, buscando o resultado mais comum dentre o obtido para as várias árvores que são geradas para construir o modelo, obtendo assim um classificador mais preciso.

A lógica de programação da floresta aleatória é a mesma dos dois modelos implementados anteriormente

```python
# chama a função da floresta aleatória do sklearn
classificador_floresta = ensemble.RandomForestClassifier(n_estimators=1000, n_jobs = -1)
```

### Matriz de Confusão - Multinomial Naive-Bayes

> Agora partindo para análise da matriz de confusão, onde será possível saber como o modelo desempenha em cada categoria da nossa variável target, sendo que os valores também irão variar de **0** a **1**, de acordo com o percentual de acertos e erros do modelo para cada categoria.

```python
tabela = pd.crosstab(df_resultados['Resultado Previsto NB'],df_resultados['Resultado Correto NB'], normalize='columns').round(4)
tabela
```

```python
verdadeiros_low_cost = tabela.iloc[0,0]*100
falsos_low_cost = 100 - verdadeiros_low_cost
verdadeiros_medium_cost = tabela.iloc[1,1]*100
falsos_medium_cost = 100 - verdadeiros_medium_cost
verdadeiros_high_cost = tabela.iloc[2,2]*100
falsos_high_cost = 100 - verdadeiros_high_cost
verdadeiros_very_high_cost = tabela.iloc[3,3]*100
falsos_very_high_cost = 100 - verdadeiros_very_high_cost


print('A porcentagem de verdadeiros low cost é: {:.2f}%'.format(verdadeiros_low_cost))
print('A porcentagem de falsos low cost é: {:.2f}%\n'.format(falsos_low_cost))
print('A porcentagem de verdadeiros medium cost é: {:.2f}%'.format(verdadeiros_medium_cost))
print('A porcentagem de falsos medium cost é: {:.2f}%\n'.format(falsos_medium_cost))
print('A porcentagem de verdadeiros high cost é: {:.2f}%'.format(verdadeiros_high_cost))
print('A porcentagem de falsos high cost é: {:.2f}%\n'.format(falsos_high_cost))
print('A porcentagem de verdadeiros very high cost é: {:.2f}%'.format(verdadeiros_very_high_cost))
print('A porcentagem de falsos very high cost é: {:.2f}%\n'.format(falsos_very_high_cost))
```

> Analisando a matriz de confusão gerada para o Multinomial Naive-Bayes, é possível perceber que o modelo acerta mais nas extremidades de variável target e não de forma linear para todas as categorias, demonstrando uma habilidade menor de prever celulares que são classificados como "medium cost" e "high cost".

### Validação Cruzada

> Método de validação utilizado para observar como o modelo funciona de forma generalzada. Para realização dessa validação, é utilizado o método K-Fold, que divide a base de dados K vezes, usando uma das parte do teste e o resto das partes para o treinamento da base. Esse método de validação possibilita uma observação mais fiel do desempenho do modelo no geral, além de possibilitar um score mais preciso e a obtenção da média e desvio-padrão do modelo.

```python
k_fold = model_selection.KFold(n_splits = 10, shuffle=True) 
resultado = model_selection.cross_val_score(classificador_nb, x, y, cv=k_fold)

for r in resultado:
    print(f'Acurácia: {r}')
print('')
print(f'Média - Acurácias : {resultado.mean()}')
print(f'Desvio padrão - Acurácias : {resultado.std()}')
```

> A análise da validação cruzada mostra que a acurácia do Multinomial Naive-Bayes gira em torno, de forma geral, da marca dos 50%, mostrando que o modelo realmente apresenta um certo problema quanto a sua acurácia geral. Além disso, por meio da análise do desvio padrão das acurácias é possível perceber um valor elevado na variação dos resultados para cada combinação de bases de dados, o que mostra que o modelo não apresenta uma grande consistência e sua precisão varia consideravelmente conforme mudam as bases de treinamento e teste.

```python
verdadeiros_low_cost3 = tabela3.iloc[0,0]*100
falsos_low_cost3 = 100 - verdadeiros_low_cost3
verdadeiros_medium_cost3 = tabela3.iloc[1,1]*100
falsos_medium_cost3 = 100 - verdadeiros_medium_cost3
verdadeiros_high_cost3 = tabela3.iloc[2,2]*100
falsos_high_cost3 = 100 - verdadeiros_high_cost3
verdadeiros_very_high_cost3 = tabela3.iloc[3,3]*100
falsos_very_high_cost3 = 100 - verdadeiros_very_high_cost3


print('A porcentagem de verdadeiros low cost é: {:.2f}%'.format(verdadeiros_low_cost3))
print('A porcentagem de falsos low cost é: {:.2f}%\n'.format(falsos_low_cost3))
print('A porcentagem de verdadeiros medium cost é: {:.2f}%'.format(verdadeiros_medium_cost3))
print('A porcentagem de falsos medium cost é: {:.2f}%\n'.format(falsos_medium_cost3))
print('A porcentagem de verdadeiros high cost é: {:.2f}%'.format(verdadeiros_high_cost3))
print('A porcentagem de falsos high cost é: {:.2f}%\n'.format(falsos_high_cost3))
print('A porcentagem de verdadeiros very high cost é: {:.2f}%'.format(verdadeiros_very_high_cost3))
print('A porcentagem de falsos very high cost é: {:.2f}%\n'.format(falsos_very_high_cost3))
```

> Analisando a matriz de confusão do modelo de Floresta Aleatória, é possível notar que esse modelo acerta mais nas extremidades da variável target e que a diferença e que essa diferença de desempenho gira em torno de 10%. Logo, é cabível concluir que o modelo não é extremamente consistente em classificar qualquer tipo de celular, mas sim aqueles que são mais baratos ou muito caros.

### Validação Cruzada

>  Validação cruzada para Floresta Aleatória:

```python
k_fold = model_selection.KFold(n_splits = 10, shuffle=True) 
resultado3 = model_selection.cross_val_score(classificador_floresta, x_pad, y, cv=k_fold)

for r in resultado3:
    print(f'Acurácia: {r}')
print('')
print(f'Média - Acurácias : {resultado3.mean()}')
print(f'Desvio padrão - Acurácias : {resultado3.std()}')
```

> Analisando a validação cruzada para o método da Floresta Aleatória, é possível notar que o modelo é preciso de modo geral, obtendo uma média de precisão de em torno de 90%. No entanto, para esse modelo é possível notar um desvio padrão de em torno de 3%, o que mostra que, quando se deparado com bases nunca antes vistas, repetidas vezes, o modelo apresenta uma certa inconsistência em seu percentual de acertos.
